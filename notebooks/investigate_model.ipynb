```python
root_dir = '../tmp/model_006b_weight/2022-12-06_11-3800/'
mat_list = ['min', 'max', 'mean', 'std']
data_np = [[] for _ in mat_list]

starting = True
for p in glob(f"{root_dir}/*_rmae_dict.npy"):
    # setup 
    if starting == True:
        starting = False
        sample = np.load(p, allow_pickle=True)[()]
        layer_names = list(sample.keys())

    fdata = np.load(p, allow_pickle=True)[()]
    fname = p.split('\\')[-1].rstrip('.npy')
    f,_,_ = fname.split('_')
    fold_id = int(f[-1])
    n_epochs = len(fdata[layer_names[0]])
    for ei in range(n_epochs):
        for i,m in enumerate(mat_list):
            record = [fold_id, ei]+[fdata[k][ei][m] for k in layer_names]
            data_np[i].append(record)
    # print(p)

df_arr = [pd.DataFrame(data=np.array(d), columns=['Fold', 'Epoch']+layer_names) for d in data_np]
record_df = pd.concat(df_arr, axis=1, keys=mat_list)
record_df.head(20)
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=4, cols=5, start_cell="bottom-left")

met_types = ['min', 'max', 'mean', 'std']
for r,met in enumerate(met_types):
    sub_df = record_df[met]
    for f in range(5):
        sub_sub_df = sub_df.loc[sub_df['Fold']==f, :]
        df3 = pd.melt(sub_sub_df, id_vars=['Fold', 'Epoch'], value_vars=sub_sub_df.columns[2:])
        fig.add_trace(go.Scatter(x=df3['Epoch'], y=df3['value'], mode="markers+lines", color=df3['variable']),
              row=r, col=f)
```
