```python
from openai.embeddings_utils import cosine_similarity, get_embedding
import numpy as np 
import pandas as pd 
```

```python
df_path = '/content/Sentence Description - Sheet1.csv'
data_df = pd.read_csv(df_path)
data_df.head()
```

```python
actions = data_df['Action']
sentences = data_df['Sentence 1']
```

```python
EMBEDDING_MODEL = "text-similarity-ada-001" #"text-embedding-ada-002"
sentence_embedding = get_embedding(text, engine=EMBEDDING_MODEL)

emb_dict = {}
for i, a in enumerate(actions):
    emb = np.array(get_embedding(sentences[i], engine=EMBEDDING_MODEL))
    emb_dict[a] = emb 
```

```python
import pickle
```

```python
with open('/content/ada_embs.pickle', 'wb') as pf:
    pickle.dump(emb_dict, pf)
```
