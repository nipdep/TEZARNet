```python
import pandas as pd 
import numpy as np 
```

```python
file_name = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
df = pd.read_csv(file_name).iloc[:, :-1]
df.head()
```

```python
m1 = 0
n1 = 2
df['EF_ch'] = df.apply(lambda r: r[m1:n1].mean(), axis=1)
df.head()
```

```python
m2 = 1
n2 = 3
df['EF_dis'] = df.apply(lambda r: r[m2:n2].mean(), axis=1)
df.head()
```

```python
df['Consump_dis'] = df.apply(lambda r: r[m2:n2].sum(), axis=1)
df.head()
```

```python
x = [1.3, 4.45]
df['adaptive_con_dis'] = df['Consump_dis'].apply(lambda i:x[1] if x[1] < i else i)
df.head()
```

```python
eff = 0.8
df['batteryEM'] = df.loc[:, ['EF_ch', 'EF_dis', 'adaptive_con_dis']].apply(lambda r: (r[0]/eff - r[1])*r[2], axis=1)
df.head()
```

```python
total_batteryEM = df['batteryEM'].sum()
total_batteryEM
```
