```python
from pytube import YouTube 
  
# where to save 
SAVE_PATH = "/content/drive/MyDrive/22_FYP42 - Zero-shot Explainable HAR/Datasets/video_dataset/vacuum_clean"
```

```python
import os

with open("/content/drive/MyDrive/22_FYP42 - Zero-shot Explainable HAR/Datasets/vacuum_cleaning.txt","r+") as f0:
  list_vid = f0.read().split("\n")

for vid in list_vid:
  print(f"https://www.youtube.com/watch?v={vid}")
  link=f"https://www.youtube.com/watch?v={vid}"
  
  try: 
      # object creation using YouTube
      # which was imported in the beginning 
      yt = YouTube(link) 
  except: 
      print("Connection Error") #to handle exception 
    
  # filters out all the files with "mp4" extension
  try: 
    stream = yt.streams.first()
    stream.download(SAVE_PATH) 
  except:
    pass
```
